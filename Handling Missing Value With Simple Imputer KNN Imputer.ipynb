```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data2.csv",na_values=["missing","*","?","/","$"])
df
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(missing_values=np.nan,strategy="mean")
```

```python
df.iloc[:,1:]= si.fit_transform(df.iloc[:,1:])
```

```python
df
```
